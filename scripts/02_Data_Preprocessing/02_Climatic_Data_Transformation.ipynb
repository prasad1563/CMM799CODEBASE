### 1. Import Libraries

```python
import pandas as pd
```

### 2. Load Data

```python
# Define the path to the Excel file
file_path = '../../data/source/climaticdata.csv'

# Load the Excel file
climatic_source_file = pd.read_csv(file_path)

print(climatic_source_file.head())
```

### 3. Processing
 - Split year and month

```python
climatic_source_file[["Year", "Month"]] = (climatic_source_file["YearMonth"].str.split("-", expand=True)
    .astype(int)
)
```

#### 3.3. Data Quality Check

```python
district_count_per_month = (
    climatic_source_file
    .groupby(["Year", "Month"])["District"]
    .nunique()
    .reset_index(name="District_Count")
)


month_count_per_district = (
    climatic_source_file
    .groupby("District")[["Month", "Year"]]
    .apply(lambda x: x.drop_duplicates().shape[0])
    .reset_index(name="Month_Count")
)


```

```python
##Check whether all the data are available
district_count_per_month.to_csv('../../data/processed/data_quality_check/climatic_district_count_per_month.csv',
                index=False)

##Check whether all the data are available
month_count_per_district.to_csv('../../data/processed/data_quality_check/climatic_month_count_per_district.csv',
                index=False)
```
