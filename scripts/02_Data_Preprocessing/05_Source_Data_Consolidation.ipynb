### 1. Import Libraries

```python
import pandas as pd
```

### 2. Load and Consolidate Data


```python
# Define the path to the processed files
file_path = f'../../data/processed/'

leptospirosis_processed = pd.read_csv(f"{file_path}1_leptospirosis_processed.csv")
climatic_processed = pd.read_csv(f"{file_path}2_climatic_processed.csv")
agriculture_processed = pd.read_csv(f"{file_path}3_agriculture_processed.csv")
socioeconomic_processed = pd.read_csv(f"{file_path}4_socioeconomic_processed.csv")
```

```python
climatic_processed
```

```python
consolidated_dataframe = (leptospirosis_processed.merge(climatic_processed,on=["District","Month","Year"])
                                                 .merge(agriculture_processed,on=["District","Year"])
                                                 .merge(socioeconomic_processed,on=["District","Year"])
                                                )
consolidated_dataframe
```

### Save Processed Data
 - data\processed

```python
##saved as a new csv
consolidated_dataframe[
    (consolidated_dataframe["Year"] != 2007)].drop(columns=["Avg_HH_Size"]).to_csv('../../data/processed/consolidated_dataframe.csv',
                index=False)
```
